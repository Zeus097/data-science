This is exercise of testing hypothesis on a small data sources for case- study.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
heights_dataset = pd.read_csv("heights.dat")
heights_dataset.head()

heights = np.array(heights_dataset)

plt.hist(heights_dataset, bins=20, edgecolor="black")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.title("Height Distribution")

plt.show()


# Boxplot
plt.boxplot(heights, vert=False)
plt.title("Boxplot of Heights")
plt.xlabel("Height (cm)")
plt.show()


mean = np.mean(heights) # Middle value of the distribution
std = np.std(heights, ddof=1) # Standard deviation of the distribution





# Z-test = a statistical test that checks whether a sample mean 
# differs significantly from a given value, 
# measured in standard deviations.


# t-test = a statistical test that checks whether a sample mean
#  differs significantly from a given value
#  when the population standard deviation is unknown, 
# using the t-distribution.




# Calxulate intervals
z_values = {
    "50%": 0.674,
    "90%": 1.645,
    "95%": 1.96,
}

for label, z in z_values.items():
    lower = mean - z * std
    upper = mean + z * std

    print(f"{label} confidence interval: [{lower:.2f}, {upper:.2f}]")


```
